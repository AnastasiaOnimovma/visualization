```python
def hist_for_atribute_set(df, labels, ax, bin_num=25, df2=pd.DataFrame(), 
                          ylim=None, ylabel=None, y_scale='linear'):
      
  for i, col in enumerate(labels):
    ax[i].set_yscale(y_scale)   
    ax[i].tick_params(labelsize=10)
    ax[i].set_xlabel(col)
    
    # уберем лишние границы графика
    ax[i].spines['right'].set_visible(False)
    ax[i].spines['top'].set_visible(False)
    ax[i].grid()

    ax[i].hist(df[col], bin_num, density=False, color='#377eb8', alpha=.75) # синий 
    if (not df2.empty):
      ax[i].hist(df2[col], bin_num, density=False, color='#e41a1c', alpha=.45) # красный '#e41a1c' 

  # если графиков меньше чем есть в сетке, убираем лишние оси
  len_labels = len(labels)
  len_ax = len(ax)
  if len_labels < len_ax:
    for ax in ax[len_labels : len_labels + (len_ax - len_labels)]:
      ax.axis('off')
```

```python
fig, ax = plt.subplots(nrows=3, ncols=4, sharey=True, figsize=(20, 10))

# поэкспериментируйте с масштабом оси Y, параметр y_scale ('linear, 'symlog', 'log')
labels_0 = ['Fluctuations_X_in_the_left_bearing','Fluctuations_Y_in_the_left_bearing','Fluctuations_X_in_the_right_bearing','Fluctuations_Y_in_the_right_bearing']
labels_1=['Fluctuations_X_in_the_left_bearing_smooth','Fluctuations_Y_in_the_left_bearing_smooth','Fluctuations_X_in_the_right_bearing_smooth','Fluctuations_Y_in_the_right_bearing_smooth']
labels_2=['Temperature_in_the_left_bearing','Temperature_in_the_right_bearing']

hist_for_atribute_set(df_8_1, labels_0, ax[0], bin_num=20, y_scale='linear') 
hist_for_atribute_set(df_8_1, labels_1, ax[1], bin_num=20, y_scale='linear') 
hist_for_atribute_set(df_8_1, labels_2, ax[2], bin_num=20, y_scale='linear') 

# ax[0].set_ylabel('Число записей')
plt.suptitle('Гистограммы: Вещественные параметры', fontsize=16, y=.92)
fig.tight_layout(rect=[0, .03, 1, .95])
```

```python
def boxplot_for_atribute_set(df, labels, ax, ylim=None, ylabel=None):
  flierprops = dict(marker='o', markersize=3)
  
  for i, col in enumerate(labels):
    ax.boxplot(df[col], positions=[i + 1], flierprops=flierprops) # i + 1
    ax.tick_params(labelsize=10)
    ax.set_xticks(np.arange(len(labels)) + 1, labels)
            
    if ylim:
      ax.set_ylim(ylim)

    if ylabel:
      ax.set_ylabel(ylabel)
```

```python
fig, ax = plt.subplots(sharey=True, figsize=(15, 7))
fig.set_figwidth(10)
plt.xticks(rotation=15)
boxplot_for_atribute_set(df_8_1, labels_0+labels_1, ax)
ax.set_title('Ящик с усами: вещественные параметры', fontsize=16)

# уберем лишние границы графика
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.grid(linestyle='--' )
fig.tight_layout(rect=[0, .03, 1, .95])
```

```python
def lineplot_for_atribute_set(df, labels, ax, clr='black',ylim=None, ylabel=None):
   
  for i, col in enumerate(labels):
    ax[i].set_xlabel(col)
    ax[i].plot(df[col], color=clr, linewidth=.65)
        
    # уберем лишние границы графика
    ax[i].spines['right'].set_visible(False)
    ax[i].spines['top'].set_visible(False)
    if clr=='black':
        ax[i].grid()   
```

```python
def heatmap(df, ax, title):
  corr = df.corr()

  # пропустим пары со слабой корреляцией, попробуем поэкспериментировать с этой настройкой
  df_corr = corr[(corr>=.4) | (corr<=-.4)]

  # print(f'Empty columns: {get_empty_columns_names(df_corr)}')

  mask = np.triu(np.ones_like(corr))
  sns.heatmap(
    df_corr, 
    cbar=False, 
    mask=mask,
    ax=ax, 
    annot=True, 
    cmap='coolwarm',
    vmin=-1,
    vmax=1,
  )

  ax.title.set_text(title)
  ax.tick_params(
    top=False,
    bottom=False,
    left=False,
    right=False,
  )
```

```python
corr_labels = labels_0+labels_2

fig, ax = plt.subplots(figsize=(7, 7), constrained_layout=True)
hm = heatmap(df_8_1[corr_labels], ax, 'Матрица корреляции:норма')
```
